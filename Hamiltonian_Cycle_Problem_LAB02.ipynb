```python
from collections import defaultdict
```

```python
def hamiltonian(vertices,edges):
    # create adjacency list
    adj=defaultdict(set)
    for u,v in edges:
        adj[u].add(v)
        adj[v].add(u)
    V=list(vertices)
    n=len(V)
    if n==0:
        return []
    start=V[0]
    path=[start]
    visited={start}

    def backtrack():
        if len(path)==n:
            if path[-1] in adj[start]:   # last node connects ending node to starting node
                return path+[start]
            return None

        curr=path[-1]
        for neighbor in adj[curr]:
            if neighbor not in visited:
                visited.add(neighbor)
                path.append(neighbor)
                result=backtrack()
                if result is not None:
                    return result   # true
                # Backtrack step
                path.pop()
                visited.remove(neighbor)
        # Hamiltonanin path does exists
        return None
    return backtrack()
```

```python
V=[1,2,3,4]
E=[(1,2),(2,3),(3,4),(4,1)]

result=hamiltonian(V,E)
print(result)
```
